```python
df1 = pd.read_excel('Task 1 PDR_run2 (x).xlsx')
df2 = pd.read_excel('Task1 PDR_run3 (x).xlsx')
df3 = pd.read_excel('Task2 Power2 (x).xlsx')
new1 = df1.drop([0,1,2,3])
run1 = pd.DataFrame(new1)
run2 = pd.DataFrame(df2)
run3 = pd.DataFrame(df3)

```

1) Confidence interval for average RTT

```python
# Fix comma decimals for runs
for df in [run1, run2, run3]:
    df['avg (ms)'] = df['avg (ms)'].astype(str).str.replace(',', '.', regex=False)
    df['avg (ms)'] = pd.to_numeric(df['avg (ms)'], errors='coerce')

confidence_levels = [0.90, 0.95, 0.99]
runs = {'Run 1': run1, 'Run 2': run2, 'Run 3': run3}

for run_name, df in runs.items():
    data = df['avg (ms)'].dropna()
    n = len(data)
    mean = data.mean()
    std = data.std()
    se = stats.sem(data)  # standard error

    print(f"\n{'='*45}")
    print(f"  {run_name} | n={n} | mean={mean:.4f} ms | std={std:.4f}")
    print(f"{'='*45}")

    for cl in confidence_levels:
        t_crit = stats.t.ppf((1 + cl) / 2, df=n - 1)  # t-distribution (small sample)
        margin = t_crit * se
        lower = mean - margin
        upper = mean + margin
        print(f"  {int(cl*100)}% CI: [{lower:.4f}, {upper:.4f}] ms  (±{margin:.4f})")
```
